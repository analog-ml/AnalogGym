**ðŸ”„ What is DC Sweep in SPICE?**

A DC Sweep is a type of SPICE analysis that systematically varies a DC source (voltage or current) over a defined range and records the circuit's steady-state response at each step. Itâ€™s essentially a way to perform parametric simulation on a DC source.

**âœ… Why Use a DC Sweep?**
- To study input-output characteristics (e.g., voltage transfer curves).

- To observe bias point movement across a range of supply or input voltages.

- To verify linearity, thresholds, gain, or saturation points.

- To check behavior across temperature (if temperature is swept).
